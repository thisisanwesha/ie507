```python
#we define a list of 22 colours since we need maximum 22 colours to represent  
colours=["red", "blue", "yellow", "green", "orange", "white", "brown", "black", "gold", "cyan", "magenta", "silver", "pink", "crimson", "coral", "maroon", "olive", "plum", "sepia", "turquoise", "violet", "purple"]
print(len(colours))
```

```python
#we define the sequence in which we select the route
route=['M1', 'M2', 'M4', 'M5', 'M6', 'M8', 'M10', '12', 'M13', '16', 'M17', '18', '21', '27', '37', '50', '60', '61', '62', '63', '67', '68']
```

```python
#Dictionary to store route as key and its colour(not yet assigned) as value
Assignment={'M1': '', 'M2': '', 'M4': '', 'M5': '', 'M6': '', 'M8': '', 'M10': '', '12': '', 'M13': '', '16': '', 'M17': '', '18': '', '21': '', '27': '', '37': '', '50': '', '60': '', '61': '', '62': '', '63': '', '67': '', '68': ''}
print(Assignment)
```

```python
no=len(route) #number of all routes
last=0 #variable to indicate index of last colour assigned
Assignment['M1']=colours[last]
for i in range(no):
  feasible_colours=colours.copy()
  #code for checking for intersection
  for j in range(i):
    flag=0 #flag for checking if existing colour can be used
    if(df.iat[i,j+1]!=1):
      col=Assignment[route[j]]
      c=0 #counter for number of intersections
      for k in range(i):
        if col==Assignment[route[k]]:
          if df.iat[i,k+1]==1:
            c=c+1
      if c==0:
        flag=1
        Assignment[route[i]]=col #reusing an already used colour
        break
    if flag==0 and j==i-1:
      last=last+1
      Assignment[route[i]]=colours[last] #assigning a new colour

#restricting list of colours to only those needed
feasible_colours=feasible_colours[0:last+1]  

#printing route, it's colour, and number of colours required
print(Assignment)
print(feasible_colours)
for k,v in Assignment.items():
  print("Colour of Route", k, "is", v)
print("Number of colours needed is", len(feasible_colours))
```
