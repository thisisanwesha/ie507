```
import pandas as pd
```

```
data_txtfile = np.loadtxt('lab6_ex2.txt')
print(data_txtfile)
```

```
M, N= data_txtfile.shape
print(M) #no of rows(factories)
print(N) #no of columns(location)
```

```
# create a model
model_lab6_ex2 = ConcreteModel()
```

```
row_indices = range(M)
col_indices = range(N)

print(row_indices)
print(col_indices)
```

```
#to get the contents of costs alone from the file we use the following
for row in row_indices:
  for col in col_indices:
    print('cost[',row,'][',col,']:',data_txtfile[row, col])
```

```
#declare the decision variables in the model
model_lab6_ex2.x = Var(row_indices, col_indices, domain= NonNegativeIntegers)
```

```
model_lab6_ex2.objective = Objective(expr=sum(data_txtfile[i,j]*model_lab6_ex2.x[i,j] for i in row_indices for j in col_indices),sense=minimize)
```

```
model_lab6_ex2.constraints = ConstraintList()
```

```
#adding constraints of locating one factory at each location and assigning one location for each factory
for i in range(M):
  model_lab6_ex2.constraints.add(expr=sum(model_lab6_ex2.x[i,j] for j in col_indices) == 1)

for j in range(N):
  model_lab6_ex2.constraints.add(expr=sum(model_lab6_ex2.x[i,j] for i in row_indices) == 1)
```

```
model_lab6_ex2.pprint()
```

```
!apt-get install -y -qq coinor-cbc
```

```
opt_cbc = SolverFactory('cbc')
```

```
result = opt_cbc.solve(model_lab6_ex2)
print('Solver status:', result.solver.status)
print('Solver termination condition:',result.solver.termination_condition)
```

```
# display solution
print('\nObjective = ', model_lab6_ex2.objective())

for i in row_indices:
  for j in col_indices:
    if model_lab6_ex2.x[i,j].value==1:
      print('Factory', i+1,'is to be located at Location', j+1)

print('\nConstraints')
model_lab6_ex2.constraints.display()
```

```
#changing domain of variables to continuous values
for i in row_indices:
  for j in col_indices:
    model_lab6_ex2.x[i,j].domain=NonNegativeReals
```

```
result = opt_cbc.solve(model_lab6_ex2)
print('Solver status:', result.solver.status)
print('Solver termination condition:',result.solver.termination_condition)
```

**Answer 1:**

$minimize \sum_{i=1}^n\sum_{j=1}^n C(i,j)*x_{ij} $

subject to constraints

$\sum_{i=1}^nx_{ij}=1\ \forall i \in \{1,2,....n\}$

$\sum_{j=1}^nx_{ij}=1\ \forall j \in \{1,2,....n\}$

$ x_{ij} \geq 0\ \forall\ i,j \in \{1,2,....,n\}$

where C(i,j) is the cost of locating factory i at location j. $x_{ij}$ indicates whether factory i is located at location j or not, for which the values are 1 and 0 respectively (since they are taken to be non negative integers and the first two constraints restricts them from taking values more than 1).
