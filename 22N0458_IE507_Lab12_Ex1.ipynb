```python
#open and read the file containing the list of 5-letter words
with open('freq_new.txt') as f:
  lines = f.readlines()
  print(lines)
  total=len(lines)
  print(total)
```

```python
#code for calculating frequency of letters
frequency_list=[0]*26
for i in lines:
  for j in range(5):
    frequency_list[ord(i[j])-97] += 1
```

```python
#printing frequency of letters in a tabular format
print("Letter\tFrequency\tAverage frequency(Frequency/Total)")
for i in range(26):
  print((chr(i+97)),"\t",(frequency_list[i]),"     \t",((frequency_list[i])/total))
```
