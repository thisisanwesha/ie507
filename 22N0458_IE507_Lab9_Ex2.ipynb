```python
import numpy as np
#Let us use bar plots now to understand about the data description
```

```python
#We will plot bar plots using seaborn
sns.barplot(x='mpg',y='displacement',data=df);
```

Answer 3) The general trend seen is that as mpg increases, displacement is less. High error bars are also observed in some cases.

```python
#Let us try the barplot with a different option
ax = sns.barplot(x='mpg',y='displacement',hue='model_year',data=df,estimator=np.median);
#We can adjust the positioning of the legend
ax.legend(loc='center left', title='model_year', bbox_to_anchor=(1, 0.5));
```

Answer 4) The graph is observed to be almost exponentially decreasing with very high error bars for some cars. 

```python
#To capture whether two attributes are correlated, one way is to use scatter plot

sns.scatterplot(x='mpg',y='horsepower',data=df);
```

Answer 5) There is negative correlation between the two attributes mpg and horsepower (they are inversly proportional).

```python
sns.jointplot(x='mpg', y='horsepower', data=df, kind="reg");
```

Answer 6) Joint plot has three plots: a bivariate scatter plot and two marginal histograms to show the distributions of mpg and horsepower. Both marginal histograms are right skewed and they are strongly negatively correlated.

Joint plot is a better visualization technique to understand the correlation between mpg and horsepower as we can see the univariate distirbutions of both mpg and horsepower as well as the bivariate scatter plot at the same time.

```python
#We can also use lineplot to plot the relationship between variables
sns.lineplot(x='model_year',y='horsepower',data=df);
```

Answer 7) The light blue band is the confidence interval around the dark blue line which is the estimated central tendency of our data. We can observe that overall, horsepower tends to decrease as model year increases. The confidence interval is also higher when horsepower increases with increase in model year compared to when horsepower decreases with increase in model year.

```python
#We can use box plots which would provide details about summary statistics 
sns.boxplot(x='model_year',y='displacement',hue='origin',data=df);
```

Answer 8) Median displacement is significantly higher for USA than Japan or Europe for all model years.

```python
sns.violinplot(x='model_year',y='acceleration',hue='origin',data=df);
```

Answer 9) For most of the model years, Europe has higher median acceleration than USA or Japan. The median acceleration lies between 10 and 20 for all model years. For model year 79 and 82, violin plot for Europe is longer because there is a large difference between it's maximum and minimum.

```python
sns.pairplot(df, x_vars=['displacement', 'horsepower'], y_vars=['acceleration'],
             hue='cylinders', height=5, aspect=.8, kind='reg');
```

Answer 10) There is weak negative correlation for all cylinders in horsepower vs acceleration plot.
There is weak negative correlation for cylinders 4 and 8 and weak positive correlation for other cylinders in displacement vs acceleration plot.

```python
sns.pairplot(df, 
             x_vars = ['mpg', 'cylinders', 'displacement', 'horsepower', 'acceleration'], 
             y_vars = ['mpg', 'cylinders', 'displacement', 'horsepower', 'acceleration'], 
             hue='origin', height=2.5, kind='reg');
```

Answer 11) For diagonal plots, we can observe almost normal distribution. 

For most of the off diagonal plots, datapoints are not random. And, datapoints of same origin can be found clustered in the similar region so we can classify them depedning on their origin. When we write kind = 'reg' then the off diagonal plots become regression plots. When the slope of orange, green and blue lines are positive, we say there is positive correlation and when the slope is negative we say there is negative correlation. Both can be observed in different graphs.
